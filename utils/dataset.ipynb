```python
import json
import os
import requests
import time

# 输入的基础 ID 文件路径
input_path = "cats.json"  # 你那份只有 name、catId 的 JSON 文件
output_folder = "cats_detail_json"
os.makedirs(output_folder, exist_ok=True)

# API endpoint
detail_base_url = "https://node-app-lmhzwhyv4q-as.a.run.app/api/cats/"

# 加载已有的 ID 列表
with open(input_path, "r") as f:
    cat_list = json.load(f)

# 循环获取详情数据
for cat in cat_list:
    cat_id = cat["catId"]
    try:
        res = requests.get(f"{detail_base_url}{cat_id}")
        res.raise_for_status()
        detail = res.json()

        # 保存完整 JSON 数据（包含 imageUrl）
        out_path = os.path.join(output_folder, f"{cat_id}.json")
        with open(out_path, "w") as out_file:
            json.dump(detail, out_file, indent=2)

        print(f"✅ Saved details for: {detail['name']} ({cat_id})")
        time.sleep(0.3)  # 防止触发请求频率限制

    except Exception as e:
        print(f"❌ Failed to fetch details for {cat_id}: {e}")
```

```python
image_folder = "cat_images"
os.makedirs(image_folder, exist_ok=True)

for file_name in os.listdir(output_folder):
    file_path = os.path.join(output_folder, file_name)
    with open(file_path, "r") as f:
        cat_data = json.load(f)
        img_url = cat_data.get("imageUrl")
        if img_url:
            try:
                img_data = requests.get(img_url)
                with open(os.path.join(image_folder, f"{cat_data['catId']}.jpg"), "wb") as img_file:
                    img_file.write(img_data.content)
                print(f"🖼️ Saved image for {cat_data['name']}")
            except Exception as e:
                print(f"❌ Failed to download image: {e}")
```
